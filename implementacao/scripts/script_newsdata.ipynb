```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
```

```python
#iniciando cliente mongo
client = MongoClient('192.168.1.9', 27017)

#selecionando db1
db1 = client['news']
#selecionando coleções db1
oantagonista = db1.get_collection('oantagonista')
oantagonista = pd.DataFrame(list(oantagonista.find()))

oantagonistaComentarios = db1.get_collection('oantagonistaComments')
oantagonistaComentarios = pd.DataFrame(list(oantagonistaComentarios.find()))

oantagonista.count()
```

```python
oantagonistaComentarios.head()
```

```python
#carregando coleções db1 em dataframe
estadao_db1 = pd.DataFrame(list(estadao_db1.find()))
folha_db1 = pd.DataFrame(list(folha_db1.find()))
carta = pd.DataFrame(list(carta.find()))

#carregando coleções db2 em dataframe
estadao_db2 = pd.DataFrame(list(estadao_db2.find()))
estadao_comentarios = pd.DataFrame(list(estadao_comentarios.find()))

folha_db2 = pd.DataFrame(list(folha_db2.find()))
folha_comentarios = pd.DataFrame(list(folha_comentarios.find()))

g1 = pd.DataFrame(list(g1.find()))
g1_comentarios = pd.DataFrame(list(g1_comentarios.find()))
```

```python
#concat db1 e db2 notícias estadao 
estadao = pd.concat([estadao_db1, estadao_db2], sort=False)
#removendo colunas irrelevantes
estadao = estadao.drop(columns=['_id'])
estadao.head(2)
```

```python
#removendo colunas irrelevantes
folha_db1 = folha_db1.drop(columns=['_id', 'ano','data','dia', 'mes'])
folha_db2 = folha_db2.drop(columns=['_id'])
#concat db1 e db2 notícias estadao 
folha = pd.concat([folha_db1, folha_db2], sort=False)
folha.head(2)
```

```python
#calculando timestamp carta
carta['timestamp'] = carta.data.apply(lambda x : pd.Timestamp(x).timestamp())
#removendo colunas irrelevantes
carta = carta.drop(columns=['ano','mes','dia','data','_id'])
carta.head(2)
```

```python
#removendo coluna irrelevante
g1 = g1.drop(columns=['_id'])
g1.head(2)
```

```python
#removendo colunas irrelevantes
estadao_comentarios = estadao_comentarios.drop(columns=['ThumbsDown','ThumbsUp','_id'])
estadao_comentarios.head(2)
```

```python
#removendo colunas irrelevantes
folha_comentarios = folha_comentarios.drop(columns=['ThumbsDown','ThumbsUp','_id'])
folha_comentarios.head(2)
```

```python
#removendo colunas irrelevantes
g1_comentarios = g1_comentarios.drop(columns=['ThumbsDown','ThumbsUp','_id'])
g1_comentarios.head(2)
```

```python
#salvando em novo database
news = client['news']

news.create_collection('estadao')
news.estadao.insert(estadao.to_dict('records'))

news.create_collection('folha')
news.folha.insert(folha.to_dict('records'))

news.create_collection('carta')
news.carta.insert(carta.to_dict('records'))

news.create_collection('g1')
news.g1.insert(g1.to_dict('records'))

news.create_collection('estadaoComentarios')
news.estadaoComentarios.insert(estadao_comentarios.to_dict('records'))

news.create_collection('folhaComentarios')
news.folhaComentarios.insert(folha_comentarios.to_dict('records'))

news.create_collection('g1Comentarios')
news.g1Comentarios.insert(g1_comentarios.to_dict('records'))
```
